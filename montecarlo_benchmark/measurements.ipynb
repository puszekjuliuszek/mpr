```python
def format_to_scientific_string(num):
    exponent = len(str(num)) - 1  # Calculate the exponent
    base = num / (10 ** exponent)   # Calculate the mantissa
    return f"{int(base)} × 10^{exponent}"

def adjust_df_for_plotting(df):
    resultdf = df.copy(deep=True)
    resultdf["czas wykonania 20"] = resultdf["czas wykonania 20"] / 20
    return resultdf
```

```python
def draw_df(df, pallete=sns.diverging_palette(150, 10, as_cmap=True), title= "Heatmap of Monte Carlo Execution Time"):
    df = adjust_df_for_plotting(df)
    heatmap_data = df.pivot(index="ile punktów", columns="ile wątków", values="czas wykonania 20")

    plt.figure(figsize=(10, 8))
    ax = sns.heatmap(heatmap_data, annot=True, fmt=".2f", cmap=pallete, cbar_kws={'label': 'Execution Time (s)'})
    y_labels = ["1 x 10^7", "2 × 10^7", "3 × 10^7", "4 × 10^7", "5 × 10^7", "6 × 10^7", "7 × 10^7", "8 × 10^7", "9 × 10^7", "10 × 10^7"]
    ax.set_yticks(range(len(y_labels)))  # Assuming you have as many y-labels as y-ticks
    ax.set_yticklabels(y_labels, rotation=0)  # Set the y-axis labels (no rotation for readability)
    plt.title(title)
    plt.xlabel("Number of Threads")
    plt.ylabel("Number of Points")
    plt.show()
```

```python
draw_df(cpp_df)
```

```python
diff_df = cpp_df.drop(index=[i for i in range(0, len(cpp_df), 11)])
diff_df["czas wykonania 20"]= diff_df["czas wykonania 20"] - elixir_df["czas wykonania 20"]
draw_df(diff_df, sns.diverging_palette(10, 150, as_cmap=True), "How many seconds elixir was faster")
```
